```python
import dlib
from numpy import random
import sys
import os
import cv2

output_dir = 'D:/Desktop/InfoProject/getdata/haha'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    
#size of capured photos
size = 250
```

```python
# we need the relight function to chaneg the light and contrast;
# we give default values when write function
# light controls light; noise change the contrast
def relight(img, light=1, noise=0):
    # the input img is a matrix[]
    h = img.shape[0]
    w = img.shape[1]

    # change every values on every pixel
    for i in range(h):
        for j in range(w):
            for k in range(3):  # (3) because it is colorful image
                middle = int(img[i, j, k] * light + noise)
                # cut the over 255 ones to 255 and make up the ones lower than 0 to zero
                if middle > 255:
                    middle = 255
                elif middle < 0:
                    middle = 0
                img[i, j, k] = middle  # exchange the original and processed image
    return img
```

```python
# face detector
detector = dlib.get_frontal_face_detector()
# open camera and get captures; (0) instead of video file means open the camera
camera = cv2.VideoCapture(0)

num_pics = 100

for n in range(num_pics):
    print("I'm processing No.%d\tpicture!" %n)
    # cv2.VideoCapture.read([image]) â†’ retval,image
    success, img = camera.read()

    # here we not change color model and detector the data
    rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    data = detector(rgb_img, 1)

    for i, d in enumerate(data):
        x1 = d.top() if d.top() > 0 else 0
        y1 = d.bottom() if d.bottom() > 0 else 0
        x2 = d.left() if d.left() > 0 else 0
        y2 = d.right() if d.right() > 0 else 0

        face = img[x1:y1, x2:y2]

        # we change light and contrast, both are random
        face = relight(face, random.normal(2, 0.5), random.uniform(-20, 30))

        # resize the picture
        face = cv2.resize(face, (size, size))

        # show the picture
        cv2.imshow('img', face)
        # wirte to this directory
        cv2.imwrite(output_dir + '/' + str(n + 100) + '.jpg', face)

    key = cv2.waitKey(80) & 0xff
    if key == 27:
        break
print("Completed!Awesome!You are genius")
```
